```python
#### dacon apt 
import pandas as pd
train = pd.read_csv('train.csv')
print(train.shape)
test = pd.read_csv('test.csv')
print(test.shape)
data = pd.concat([train,test],axis=0)
display(data.head())
data.shape
```
